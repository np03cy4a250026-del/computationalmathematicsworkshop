b. What are the differences between NumPy array and python list?

Data type storage: NumPy arrays are strictly homogeneous (all elements must be of the same data type), whereas Python lists can be heterogeneous (storing mixed data types like integers, strings, and floats together).

Importing modules: To use NumPy arrays, you must explicitly import the library (import numpy as np). Python lists are built-in and require no external imports.

Numerical operations: NumPy arrays support element-wise operations directly (e.g., multiplying the whole array by 2). Lists do not support this inherently; attempting to multiply a list by an integer simply repeats the list.

Memory consumption: NumPy arrays are densely packed in memory and consume significantly less memory than Python lists, making them vastly more efficient for large datasets.

Fast and convenient to use: Because NumPy is implemented in C and avoids the overhead of Python's type checking during loops, operations on arrays are executed much faster than equivalent operations on Python lists.

Difference: For 2D arrays, both perform matrix multiplication. However, np.matmul() handles higher-dimensional arrays according to the rules of matrix broadcasting and does not allow multiplication by a scalar. np.dot() handles higher dimensions differently (sum product over the last axis of $a$ and the second-to-last of $b$) and does allow scalar multiplication.

Result of np.dot(5, arr1): This performs scalar multiplication, multiplying every element in arr1 by 5.Result: [[ 5 10 15], [10 15 20], [10 15 20]].

```python
#Example 7:
#Create two vectors in 2-dimensional (2D) space and plot them using Matplotlib.
#Solution: A vector in 2D space can created as a 1D ndarray with two elements as:
V = np.array([1,1])
W = np.array([-2.0,-1])
#To plot the above vectors, we will use the quiver() function from the pyplot module of Matplotlib
#library. The quiver() function is used to create 2D field or velocity plots and is perfect for our purpose
#of vector plotting. Here is an example of plotting the above vectors.
import numpy as np
import matplotlib.pyplot as plt
""" Problem 2: Create 2-dimensional position vectors and plot them. """
origin = np.array([0,0])
V = np.array([1,1])
W = np.array([-2.0,-1])
fig, ax = plt.subplots()
# Add vectors V and W to the plot
ax.quiver(origin[0], origin[1], V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(origin[0], origin[1], W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.text(origin[0]-0.1, origin[1]+0.1, "O")
ax.text(V[0], V[1], "v")
ax.text(W[0]-0.125, W[1]-0.1, "w")
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.grid()
plt.show()
```

a. Create three vectors in 3D space and plot them using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])
origin = np.array([0, 0, 0])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(*origin, *U, color='r', label='U')
ax.quiver(*origin, *V, color='g', label='V')
ax.quiver(*origin, *W, color='b', label='W')

ax.set_xlim([0, 1.5])
ax.set_ylim([0, 1.5])
ax.set_zlim([0, 1.5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.legend()
plt.show()
```

b. Create two vectors in 2D space, add them to create a third vector, and plot all three.

```python
import numpy as np
import matplotlib.pyplot as plt

U = np.array([-0.5, 0.5])
V = np.array([1, 1])
W = U + V
origin = np.array([0, 0])

fig, ax = plt.subplots()
ax.quiver(*origin, *U, angles='xy', scale_units='xy', scale=1, color='r', label='U')
ax.quiver(*origin, *V, angles='xy', scale_units='xy', scale=1, color='g', label='V')
ax.quiver(*origin, *W, angles='xy', scale_units='xy', scale=1, color='b', label='W=U+V')

ax.set_xlim([-1, 2])
ax.set_ylim([0, 2])
ax.grid()
plt.legend()
plt.show()
```

c. Write a function to calculate the magnitude of a vector.

```python
def calculate_magnitude(vector):
    return np.linalg.norm(vector)

U = np.array([1.5, 1.5, 2.0])
print("Magnitude:", calculate_magnitude(U))
```

d. Write a function to calculate the direction cosines of the vector.

```python
def direction_cosines(vector):
    magnitude = np.linalg.norm(vector)
    return vector / magnitude

O = np.array([1.5, 1.5, 2.0])
print("Direction Cosines:", direction_cosines(O))
```

e. Write a function to calculate Scalar (dot) product of two vectors.

```python
def dot_product(v1, v2):
    return np.dot(v1, v2)

a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])
print("Dot Product:", dot_product(a, b))
```
