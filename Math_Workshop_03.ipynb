9. Plot the trigonometric COSINE function in interval [−3𝜋,3𝜋], i.e., 𝑦 = cos (𝑥), −3𝜋 ≤ 𝑥 ≤3𝜋

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3*np.pi, 3*np.pi, 100)

y = np.cos(x)

plt.plot(x, y)

plt.xlabel("x")
plt.ylabel("cos(x)")
plt.title("y = cos(x), -3π ≤ x ≤ 3π")

plt.grid()

plt.show()
```

10. Plot a circle with center at (a, b) and radius r. The values of a=b=2, and r=4, using a while loop.

```python
import matplotlib.pyplot as plt
import math

a = 2
b = 2
r = 4

x = []
y = []

theta = 0
end = 2 * math.pi
step = 0.1

while theta <= end:
    x.append(r * math.cos(theta) + a)
    y.append(r * math.sin(theta) + b)
    theta += step

plt.plot(x, y)

plt.plot(a, b, 'o')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Circle with center (2,2) and radius 4")

plt.grid()
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```

11. Plot the above circle and the ellipse on the same axes. This is give an idea about the intersection points of the two conic sections.

```python
import matplotlib.pyplot as plt
import math

a = 2
b = 2
r = 4

x_circle = []
y_circle = []

theta = 0
end = 2 * math.pi
step = 0.1

while theta <= end:
    x_circle.append(r * math.cos(theta) + a)
    y_circle.append(r * math.sin(theta) + b)
    theta += step

a_e = 3
b_e = 5
center = [2, 2]

x_ellipse = []
y_ellipse = []

theta = 0

while theta <= end:
    x_ellipse.append(a_e * math.cos(theta) + center[0])
    y_ellipse.append(b_e * math.sin(theta) + center[1])
    theta += step

plt.plot(x_circle, y_circle, label="Circle (r=4)")
plt.plot(x_ellipse, y_ellipse, label="Ellipse (a=3, b=5)")

plt.plot(2, 2, 'o')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Circle and Ellipse on Same Axes")

plt.grid()
plt.legend()

plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```

12. Plot a Parabola (𝑦 − 𝑝) = 4𝑎(𝑥 − 𝑞)^2 with vertex at (𝑝,𝑞). The values of p, q, and a can be chosen appropriately.

```python
import matplotlib.pyplot as plt
import numpy as np

p = 0
q = 0
a = 0.5

x = np.linspace(-5, 5, 100)

y = p + 4*a*(x - q)**2

plt.plot(x, y)

plt.plot(q, p, 'o')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Parabola with vertex at (0,0)")

plt.grid()

plt.show()
```
