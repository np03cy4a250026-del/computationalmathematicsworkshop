Task-1:
Create a scatter graph of 10 random x and y values between 0 and 50. Label the axis to show
x – Time taken in seconds
y – Distance reached
Title “Buzzer Game Results”
One you have a working program, add the following code to your program;
colors=["red","green","blue","yellow","pink","orange","purple","brown"
,"cyan","magenta"]
And add to your scatter() call the ‘c=colors’ argument like so;

What happens?

```python
import matplotlib.pyplot as plt
import random

# Generate 10 random x and y values between 0 and 50 [cite: 40]
x = [random.randint(0, 50) for _ in range(10)]
y = [random.randint(0, 50) for _ in range(10)]

# Initial plot
plt.scatter(x, y)
plt.xlabel("Time taken in seconds") # [cite: 41]
plt.ylabel("Distance reached")      # [cite: 42]
plt.title("Buzzer Game Results")    # [cite: 43]

# Task 1 expansion: Adding colors [cite: 44, 45, 46]
colors = ["red", "green", "blue", "yellow", "pink", "orange", "purple", "brown", "cyan", "magenta"]
plt.scatter(x, y, c=colors) #
plt.show()
```

==> What happens? When you add the c=colors argument, Matplotlib maps each color in the list to the corresponding point in the scatter plot based on its index. Instead of all points being the same color, each individual point will now be displayed in its own unique color from the list provided.

Task 2:
Following data shows the change in population of different animals from 2017 to 2022.
a. Create a pie chart for the population of each year
b. Can we show all the pie chart on the single pane? Write the code.

```python
import matplotlib.pyplot as plt

# 1. Define the categories and years
animals = ['Bears', 'Dolphins', 'Whales']
years = [2017, 2018, 2019, 2020, 2021, 2022]

# 2. Store the data for each year
data = {
    2017: [8, 150, 80],
    2018: [54, 77, 54],
    2019: [93, 32, 100],
    2020: [116, 11, 76],
    2021: [137, 6, 93],
    2022: [184, 1, 72]
}

# 3. Create a figure with a grid of subplots (2 rows, 3 columns)
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Flatten the axes array from 2D to 1D to make it easier to loop through
axes = axes.flatten()

# 4. Loop through the years and plot a pie chart on each subplot
for i, year in enumerate(years):
    axes[i].pie(data[year], labels=animals, autopct='%1.1f%%', startangle=140)
    axes[i].set_title(f"Population in {year}")

# 5. Add a main title and adjust the layout so they don't overlap
plt.suptitle("Animal Population Changes (2017-2022)", fontsize=16)
plt.tight_layout()

# 6. Display the charts
plt.show()
```

Task 3:
Construct a pie chart to visually display the favorite fruits of the students in a class based on the
given data: Mango - 45; Orange - 30; Plum - 15; Pineapple - 30; Melon – 30


```python
import matplotlib.pyplot as plt

# Data from task [cite: 69, 70]
fruits = ['Mango', 'Orange', 'Plum', 'Pineapple', 'Melon']
counts = [45, 30, 15, 30, 30]

plt.pie(counts, labels=fruits, autopct='%1.1f%%')
plt.title("Favorite Fruits of Students")
plt.show()
```

```python
import matplotlib.pyplot as plt

colors_labels = ["Red", "Silver", "Blue", "White", "Green"]
counts = [123, 230, 78, 193, 12]

# 4a & 4d: Bar chart with specific colors [cite: 96, 99, 100]
# Use actual color names recognized by Matplotlib (Silver is replaced by lightgray/silver)
car_colors = ["red", "silver", "blue", "whitesmoke", "green"]

plt.bar(colors_labels, counts, color=car_colors, width=0.3) # 4b: width=0.3
plt.title("Number of Cars Seen by Color")
plt.xlabel("Car Color")
plt.ylabel("Number Seen")
plt.show()

# 4c: Horizontal bar chart
plt.barh(colors_labels, counts, color=car_colors)
plt.title("Horizontal Car Data")
plt.show()
```

Effects observed:


width=0.3: Reduces the width of the bars, creating more space between them.


plt.barh(): Flips the axes, turning vertical bars into horizontal bars.

```python
import matplotlib.pyplot as plt

# Using the car data [cite: 101]
colors_labels = ["Red", "Silver", "Blue", "White", "Green"]
counts = [123, 230, 78, 193, 12]

# 4f: Testing explode, shadow, legend, and custom colors [cite: 102, 103]
myexplode = [0.3, 0, 0, 0, 0] # Explode the first slice (Red) [cite: 103]
mycolors = ["magenta", "silver", "cyan", "red", "green"] # Custom colors [cite: 103]

plt.pie(counts, labels=colors_labels, explode=myexplode, shadow=True, colors=mycolors) # [cite: 103]
plt.legend(title="Car Colours:") # [cite: 103]
plt.title("Car Color Distribution") # [cite: 101]
plt.show()
```
